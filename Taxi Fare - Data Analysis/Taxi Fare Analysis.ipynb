## **SQL**

##### 1. Apa saja nama perusahaan yang terdapat dalam data?

```python
# Mencari nilai unik dari kolom company

df_1 = client.query('''
SELECT DISTINCT company as `Nama Perusahaan`
FROM `ftds-hacktiv8-project.phase0_ftds_005_bsd.taxi_fare`
''').to_dataframe()

df_1
```

##### 2. Berapa nilai minimum dan maksimum dari waktu pengambilan data?

```python
# Mencari nilai min dan max dari waktu pengambilan data

df_2 = client.query('''
SELECT MIN(trip_start_timestamp) as `Nilai Minimum`,
       MAX(trip_start_timestamp) as `Nilai Maximum`
FROM `ftds-hacktiv8-project.phase0_ftds_005_bsd.taxi_fare`
''').to_dataframe()

df_2
```

Nilai minimum dari seluruh perjalanan adalah pada tanggal 2 Januari 2021 pukul 11.00, sementara nilai maximum adalah pada tanggal 31 Desember 2022 pukul 14.15.

##### 3. Bandingkan jumlah banyaknya perjalanan antara seluruh perusahaan pada tahun 2022?

```python
# Menampilkan banyaknya perjalanan berdasarkan perusahaan pada tahun 2022

df_3 = client.query('''
SELECT company as `Nama Perusahaan`,
       (COUNT(unique_key)) as `Jumlah Perjalanan`
FROM `ftds-hacktiv8-project.phase0_ftds_005_bsd.taxi_fare`
WHERE EXTRACT(YEAR FROM trip_start_timestamp) = 2022
GROUP BY 1
ORDER BY 2 DESC
''').to_dataframe()

df_3
```

Berdasarkan data di atas, perusahaan yang memiliki banyaknya perjalanan terbanyak pada tahun 2022 adalah Synctaxi Inc. dengan total 1341 perjalanan, diikuti oleh Federway Int. dengan total  1059 perjalanan, diikuti oleh Deluc Porter dengan total 1050 perjalanan, dan terakhir adalah Rocket Cab Auto dengan total 77 perjalanan. Sementara itu, dari tabel di atas, terlihat bahwa perusahaan Deluc Porter tidak memiliki perjalanan pada tahun 2022.

##### 4. Perusahaan mana yang memiliki tarif perjalanan per km termurah?

```python
# Menghitung tarif perjalanan per km termurah

df_4 = client.query('''
SELECT company as `Nama Perusahaan`,
       SUM(fare/trip_km) as `Tarif Perjalanan per KM`
FROM `ftds-hacktiv8-project.phase0_ftds_005_bsd.taxi_fare`
GROUP BY 1
ORDER BY 2 ASC
''').to_dataframe()

# Menghitung rata-rata tarif perjalanan per km termurah berdasarkan company

df_4b = client.query('''
SELECT company as `Nama Perusahaan`,
       AVG(fare/trip_km) as `Tarif Perjalanan per KM`
FROM `ftds-hacktiv8-project.phase0_ftds_005_bsd.taxi_fare`
GROUP BY 1
ORDER BY 2 ASC
''').to_dataframe()

# df_4 = client.query('''
# SELECT company as `Nama Perusahaan`,
#        IEEE_DIVIDE(fare,trip_km) as `Tarif Perjalanan per KM`
# FROM `ftds-hacktiv8-project.phase0_ftds_005_bsd.taxi_fare`
# WHERE 'Tarif Perjalanan per KM' != '0.0'
# ORDER BY 2 ASC
# ''').to_dataframe()

df_4
df_4b
```

Pada data di atas, tarif perjalanan per KM termurah dari seluruh perjalanan diraih oleh Rocket Cab Auto. Perusahaan selanjutnya yang memiliki rata-rata perjalanan per km termurah adalah Atomic Inc., diikuti oleh Synctaxi Inc., diikuti oleh Deluc Porter, dan terakhir adalah Federway Int.

## **PANDAS**

```python
# Data loading

df = client.query('''
              SELECT *
              FROM `ftds-hacktiv8-project.phase0_ftds_005_bsd.taxi_fare`
              LIMIT 3000
              ''').to_dataframe()

# Menampilkan dataframe
df
```

##### 1. Buat kolom baru, dengan nama Keterangan_Waktu. Kolom ini berisikan label yang menunjukkan representasi keterangan waktu berdasarkan jam dengan ketentuan berikut:

```python
# Mengecek nilai unik jam dari kolom trip_start_timestamp

df['trip_start_timestamp'].dt.hour.unique()
```

```python
# Menyimpan nilai kosong pada variabel
time = []

# Melakukan proses looping untuk menentukan kondisi
for i in df['trip_start_timestamp'].dt.hour:
  if i in (4, 5, 6, 7, 8, 9):
    time.append('Pagi')
  elif i in (10, 11, 12, 13, 14, 15):
    time.append('Siang')
  elif i in (16, 17, 18):
    time.append('Sore')
  elif i in (19, 20, 21, 22, 23):
    time.append('Malam')
  elif i in (0, 1, 2, 3):
    time.append('Dini Hari')

# Membuat kolom baru dengan variabel sebelumnya
df['Keterangan_Waktu'] = time

# Menampilkan dataframe terbaru
df
```

##### 2. Berapa total jarak perjalanan yang dicapai perusahaan Federway Int. pada tahun 2022? (Gunakan pandas query untuk menjawab soal no. 2)

```python
# Menghitung total jarak perjalanan perusahaan

total = df[(df['company'] == 'Federway Int.') &
           (df['trip_start_timestamp'].dt.year == 2022)]['trip_km'].sum().round(2)

print(f"Total jarak perjalanan yang dicapai oleh perusahaan Federway Int. pada tahun 2022 adalah {total} km")
```
